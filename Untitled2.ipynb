```python
import pandas as pd
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/Wanman-Gao/212B-Spring/refs/heads/main/thyroidDF.csv")
```

```python
#Check for missing percentage
import seaborn as sns
import matplotlib.pyplot as plt
missing_counts = data.isnull().sum()
missing_percentage = (missing_counts / len(data)) * 100
plt.figure(figsize=(12, 8))
sns.barplot(x=missing_percentage.values, y=missing_percentage.index)
plt.title('Missing Data Percentage by Feature')
plt.xlabel('Percentage of Missing Values')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
from sklearn.impute import SimpleImputer

# Select numeric columns only
numeric_cols = data.select_dtypes(include=["float64", "int64"]).columns

# Initialize median imputer
imputer = SimpleImputer(strategy="median")

# Fit-transform and assign to new DataFrame
df_median = data.copy()  # First make a full copy
df_median[numeric_cols] = imputer.fit_transform(data[numeric_cols])
```

```python

df_1 = data.replace({'t': 1, 'f': 0})

for col in df_1.select_dtypes(include="object").columns:
    try:
        df_1[col] = df_1[col].astype(int)
    except:
        pass
```

```python
# only keep numeric values
numeric_df = df_1.select_dtypes(include=["int64", "float64"])

# calculate corrlate matrix
corr_matrix = numeric_df.corr()

# draw heatmap
plt.figure(figsize=(16, 14))
sns.heatmap(corr_matrix, cmap='coolwarm', annot=True, fmt='.2f', square=True)
plt.title("Heatmap of All Numeric Variables", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
columns_to_drop = [
    "TT4",         # Highly correlated with FTI
    "TBG",         # Too many missing values, redundant with FTI
    "patient_id",  # No predictive value
    "TSH_measured", "T3_measured", "TT4_measured",
    "T4U_measured", "FTI_measured", "TBG_measured"  # Measurement flags, not actual data
]

df = data.drop(columns=columns_to_drop, errors="ignore")
```

```python
# Define mapping dictionary
mapping = {
    "-": "Negative",
    "A": "Hyperthyroid", "AK": "Hyperthyroid", "B": "Hyperthyroid",
    "C": "Hyperthyroid", "C|I": "Hyperthyroid", "D": "Hyperthyroid", "D|R": "Hyperthyroid",
    "E": "Hypothyroid", "F": "Hypothyroid", "FK": "Hypothyroid", "G": "Hypothyroid",
    "GK": "Hypothyroid", "GI": "Hypothyroid", "GKJ": "Hypothyroid", "H|K": "Hypothyroid"
}

# Filter out any rows whose 'target' is not in the mapping keys
df_median = df_median[df_median["target"].isin(mapping.keys())]

# Apply the mapping
df_median["target"] = df_median["target"].map(mapping)
```
